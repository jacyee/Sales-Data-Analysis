```python
# load dataset
import pandas as pd
df = pd.read_csv('sales-data.csv')
df.sample(10)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
print(df.shape)
```

```python
base_color = sb.color_palette()[2]
# create sorted bar chart by store type frequency
gen_order = df['Store'].value_counts().index
```

```python
sb.countplot(data = df,x='Store',color = base_color, order = gen_order);
plt.xticks(rotation = 90);
```
